```python
# 导包、参数定义
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] #黑体
plt.rcParams['font.size'] = 15 #黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时'-'显示为方块的问题

# 矩阵
M=[
    [10,0,9,6,0],
    [0,14,0,7,0],
    [5,0,0,3,0],
    [1,8,0,0,8],
    [0,4,0,0,12],
    [8,0,5,0,0],
    [0,2,0,0,6]
]
# U：左奇异矩阵；D：特征向量矩阵；Vt：右奇异矩阵
U, D,Vt = np.linalg.svd(M)

# 绘图
lable1=['Python','Java','爬虫','JavaScript','HTML','数据分析','网页开发']
lable2=['A','B','C','D','E']
X1=U[:,:3][:,1] # 取左奇异矩阵的第二维
Y1=U[:,:3][:,2] # 取左奇异矩阵的第三维
X2=Vt[:3,:][1,:] # 取右奇异矩阵的第二维
Y2=Vt[:3,:][2,:] # 取右奇异矩阵的第三维
plt.scatter(X1,Y1,s=30,label='词语')
plt.scatter(X2,Y2,s=30,label='文档')
for i,(x, y) in enumerate(zip(X1, Y1)):
    if lable1[i] == '数据分析':
        plt.annotate('%s'%(lable1[i]), xy=(x,y),xytext = (0, 0), textcoords = 'offset points',ha = 'right', va = 'top')
    elif lable1[i] == '网页开发':
        plt.annotate('%s'%(lable1[i]), xy=(x,y),xytext = (0, 0), textcoords = 'offset points',ha = 'right', va = 'top')
    else:
        plt.annotate('%s'%(lable1[i]), xy=(x,y),xytext = (0, 0), textcoords = 'offset points',ha = 'left', va = 'top')
for i,(x, y) in enumerate(zip(X2, Y2)):
    plt.annotate('%s'%(lable2[i]), xy=(x,y),xytext = (0, 0), textcoords = 'offset points',ha = 'right', va = 'top')
# 画圆
plt.scatter(-0.2,-0.5,marker='o',edgecolors='r',s=10000,alpha=0.1)
plt.scatter(0.5,0.1,marker='o',edgecolors='b',s=25000,alpha=0.1)
plt.scatter(-0.15,0.6,marker='o',edgecolors='g',s=5000,alpha=0.1)
plt.legend(loc='lower right')
plt.title('词-文档空间表示')
plt.xlabel('第二维')
plt.ylabel('第三维')
plt.tight_layout()
# 保存图片
plt.savefig('./data/pictures/词-文档空间表示.png')
plt.show()
```
